```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Upload the file
from google.colab import files
uploaded = files.upload()  # Prompts to upload Excel file

# Load the file into a Pandas DataFrame
Dataset = list(uploaded.keys())[0]  # Gets the uploaded file name
df = pd.read_excel(Dataset)
```

```python
print(df)
```

```python
# Define the mapping of grades to numerical values
grade_mapping = {
    "O": 10, "A+": 9, "A": 8, "B+": 7, "B": 6,
    "C": 5, "D": 4, "F": 3, "P": 4, "DET" : 0, "DB" :0, "AB":0
}
```

```python
# Replace grades in columns 5, 9, and 12 (index 4, 8, and 11)
columns_to_replace = [5, 6, 7, 8, 9, 10, 11, 12]
for col in columns_to_replace:
    df.iloc[:, col] = df.iloc[:, col].map(grade_mapping).fillna(df.iloc[:, col])
```

```python
print(df)
```

```python
#the horizontal mean of values from index 5,6,7,8,9,10,11,12
# then same the mean value to index 15

# Calculate the horizontal mean for columns 5-12 (index 4-11)
df['Horizontal_Mean'] = df.iloc[:, 5:13].mean(axis=1)

# Assign the mean value to index 15 (column 16)
df.insert(15, 'Mean_at_15', df['Horizontal_Mean'])

# Remove the temporary 'Horizontal_Mean' column if you don't need it
df = df.drop('Horizontal_Mean', axis=1)

df
```

```python
#the mean value at 15 index give class value at 16 index
# mean 9 to 10 class 1
# mean 6 to 8 class 2
# mean 4 to 5 class 3
# mean 0 to 3 class 4

import numpy as np
# ... (Your existing code) ...

# Define the conditions and corresponding class values
conditions = [
    (df['Mean_at_15'] >= 9),
    (df['Mean_at_15'] >= 6) & (df['Mean_at_15'] < 9),
    (df['Mean_at_15'] >= 4) & (df['Mean_at_15'] < 6),
    (df['Mean_at_15'] >= 0) & (df['Mean_at_15'] < 4)
]
choices = [1, 2, 3, 4]

# Assign the class value based on the mean value at index 15
df['Class_at_16'] = np.select(conditions, choices, default=np.nan)

# ... (Rest of your code) ...
```

```python
# Save the modified DataFrame to a new Excel file
new_file_path = 'modified_data_expirement_2_file.xlsx'  # Specify the new file path
df.to_excel(new_file_path, index=False)  # Save to Excel without index

# Download the new file
files.download(new_file_path)
```

```python
#total number of students that fall under class 1,2,3,4

# Count the number of students in each class
class_counts = df['Class_at_16'].value_counts()

# Print the counts for each class
print("Number of students in each class:")
print(class_counts)

# Alternatively, you can access the count for a specific class
class1_count = class_counts.get(1, 0)  # Get count for class 1, default to 0 if not found
class2_count = class_counts.get(2, 0)
class3_count = class_counts.get(3, 0)
class4_count = class_counts.get(4, 0)

print(f"Class 1: {class1_count}")
print(f"Class 2: {class2_count}")
print(f"Class 3: {class3_count}")
print(f"Class 4: {class4_count}")
```
