• You are working with a dataset containing temperatures recorded over a month. Some of the temperature readings are missing and are
• répresented as np.nan. Write a NumPy program to:
• Replace all missing values with the average temperature (excluding np.nan values).
• Count the number of missing values before replacement.

```python
import numpy as np

temperatures = np.array([30.5, 32.0, np.nan, 31.2, 29.8, np.nan, 33.1, 30.0, np.nan, 28.9])

missing_count = np.isnan(temperatures).sum()
print(f"Number of missing values: {missing_count}")

mean_temperature = np.nanmean(temperatures)

temperatures[np.isnan(temperatures)] = mean_temperature

print("Updated temperature dataset:")
print(temperatures)
```

1. Sales Performance Analysis
• Scenario:
• A retail company wants to analyze monthly sales performance and visualize the sales trend over the year.

```python
import pandas as pd
import matplotlib.pyplot as plt

months = np.array(["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
sales = np.array([50, 55, 53, 60, 65, 58, 70, 75, 72, 68, 80, 85])  # Sample sales data

total_sales = np.sum(sales)
average_sales = np.mean(sales)
highest_sales = np.max(sales)
lowest_sales = np.min(sales)
best_month = months[np.argmax(sales)]
worst_month = months[np.argmin(sales)]

print(f"Total Sales: ${total_sales}K")
print(f"Average Monthly Sales: ${average_sales:.2f}K")
print(f"Highest Sales: ${highest_sales}K in {best_month}")
print(f"Lowest Sales: ${lowest_sales}K in {worst_month}")

plt.figure(figsize=(10, 5))
plt.plot(months, sales, marker='o', linestyle='-', color='b', label="Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales (in $1000s)")
plt.title("Monthly Sales Performance")
plt.grid(True)
plt.legend()
plt.show()
```

Matrix Operations- Image Processing Simulation
• You are processing an image represented as a 5x5 grayscale pixel matrix (values between 0 and 255).
- Perform the following operations using NumPy:
• Normalize the matrix values between 0 and 1.
• Compute the average pixel intensity.

```python

image_matrix = np.array([
    [120, 135, 200, 255, 100],
    [60, 75, 180, 220, 90],
    [30, 50, 160, 200, 80],
    [10, 25, 140, 190, 70],
    [5, 15, 120, 180, 60]
])

normalized_matrix = image_matrix / 255.0

average_intensity = np.mean(normalized_matrix)

print("Original Grayscale Image Matrix:")
print(image_matrix)

print("\nNormalized Image Matrix:")
print(np.round(normalized_matrix, 2))

print(f"\nAverage Pixel Intensity: {average_intensity:.4f}")
```

You have a NumPy array of 1,000,000 random numbers between 1 and 1000.
Write an efficient NumPy function to:
• Find the sum of all elements.
• Compute the standard deviation of the array.
• Find the maximum and minimum values.

```python
random_numbers = np.random.randint(1, 1001, size=1_000_000, dtype=int)

total_sum = np.sum(random_numbers)
std_dev = np.std(random_numbers)
max_value = np.max(random_numbers)
min_value = np.min(random_numbers)

print(f"Sum of all elements: {total_sum}")
print(f"Standard Deviation: {std_dev}")
print(f"Maximum Value: {max_value}")
print(f"Minimum Value: {min_value}")
```
