```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Upload the file
from google.colab import files
uploaded = files.upload()  # Prompts you to upload your Excel file

# Load the file into a Pandas DataFrame
Dataset = list(uploaded.keys())[0]  # Gets the uploaded file name
df = pd.read_csv(Dataset)
```

```python
print("DataFrame Information:")
print(df.info())
print("DataFrame description:")
print(df.describe())
print("null values in the dataset")
print(df.isnull().sum())
```

```python
# Display the table
print(df.to_string(index=False))
```

```python
# Load the dataset
#file_path = "/content/diabetes.csv"
#df = pd.read_csv(file_path)
# Set Seaborn style
sns.set_style("whitegrid")

# 1. Histogram for feature distribution (Fixed)
plt.figure(figsize=(3, 1.5))
features = ['Glucose', 'BMI', 'Age']
colors = ['blue', 'green', 'red']

for i, feature in enumerate(features):
    plt.hist(df[feature].dropna(), bins=20, alpha=0.5, color=colors[i], label=feature)

plt.title("Distribution of Glucose, BMI, and Age", fontsize=16)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
# 2. Box Plot to detect outliers
plt.figure(figsize=(4, 2))
sns.boxplot(data=df[['Glucose', 'BloodPressure', 'BMI', 'Age']], palette="Set2")
plt.title("Box Plot of Numerical Features", fontsize=16)
plt.show()
```

```python

# 3. Pair Plot to see relationships between features
sns.pairplot(df[['Glucose', 'BloodPressure', 'BMI', 'Age', 'Outcome']], hue='Outcome', palette='coolwarm')
plt.show()
```

```python
# 4. Count Plot for Diabetes Outcome Distribution
plt.figure(figsize=(6, 4))
sns.countplot(x=df['Outcome'], palette="viridis")
plt.title("Diabetes Outcome Distribution", fontsize=16)
plt.xlabel("Outcome (0 = No Diabetes, 1 = Diabetes)")
plt.ylabel("Count")
plt.show()
```

```python
# 5. Correlation Heatmap
plt.figure(figsize=(5, 3))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap", fontsize=16)
plt.show()
```
